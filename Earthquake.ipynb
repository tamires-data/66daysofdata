```python
mask = data.isnull()
total = mask.sum()
percent = 100*mask.mean()
mask
total
percent
```

```python
missing_data = pd.concat([total, percent], axis=1,join='outer',
 keys=['count_missing', 'perc_missing'])
missing_data
```

```python
missing_data.sort_values(by='perc_missing', ascending=False, inplace=True)
missing_data
```
